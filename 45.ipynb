Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

- Triangle - $T_n = \dfrac{n(n+1)}{2} = 1, 3, 6, 10, 15, ...$
- Pentagonal - $P_n = \dfrac{n (3n−1)}{2} =  1, 5, 12, 22, 35, ... $
- Hexagonal	- $H_n=n(2n−1) = 1, 6, 15, 28, 45, ...$
```

It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.

````python
limit = 1000000
triangles = [n * (n+1)/2 for n in range(1,limit)]
pentagonals = [n*(3*n-1)/2 for n in range(1,limit)]
hexagonals = [n*(2*n-1) for n in range(1,limit)]
````

````python
triangles[284], pentagonals[164], hexagonals[142]
````

````python
triangles[284] == pentagonals[164] == hexagonals[142]
````

````python
set(triangles) & set(pentagonals) & set(hexagonals)
````
