The product of these numbers is 26 × 63 × 78 × 14 = 1788696.

What is the greatest product of four adjacent numbers in the same direction (up, down, left, right, or diagonally) in the 20×20 grid?

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("11.csv", header=None, delim_whitespace=True)
data = df.to_numpy()
data.shape
```

```python
max_prod = 0
n = 4
max_i = data.shape[0] # max rows
max_j = data.shape[1] # max columns

# try all numbers up
for i in range(0 + n, max_i):
    for j in range(0, max_j):
        new_prod = data[i,j] 
        for k in range(-1, -n, -1):
            new_prod *= data[i+k, j]
        if new_prod > max_prod:
            max_prod = new_prod
            print("Found new max up", max_prod)

# try all numbers down
for i in range(0 , max_i - n):
    for j in range(0, max_j):
        new_prod = data[i,j] 
        for k in range(1, n):
            new_prod *= data[i+k, j]
        if new_prod > max_prod:
            max_prod = new_prod
            print("Found new max down", max_prod)

# try all numbers left
for i in range(0 , max_i):
    for j in range(n, max_j):
        new_prod = data[i,j] 
        for k in range(-1, -n, -1):
            new_prod *= data[i, j+k]
        if new_prod > max_prod:
            max_prod = new_prod
            print("Found new max left", max_prod)

# try all numbers right
for i in range(0 , max_i):
    for j in range(0, max_j - n):
        new_prod = data[i,j] 
        for k in range(1, n):
            new_prod *= data[i, j+k]
        if new_prod > max_prod:
            max_prod = new_prod
            print("Found new max right", max_prod)

# try all numbers / diagonal
for i in range(0 , max_i):
    for j in range(0, max_j):
        try:
            new_prod = data[i,j] 
            for k in range(1, n):
                new_prod *= data[i+k, j-k]
            if new_prod > max_prod:
                max_prod = new_prod
                print("Found new max / diagonal", max_prod)
        except:
            print("Failed / diagonal for indices", i, j)

# try all numbers \ diagonal
for i in range(0 , max_i):
    for j in range(0, max_j):
        try:
            new_prod = data[i,j] 
            for k in range(1, n):
                new_prod *= data[i+k, j+k]
            if new_prod > max_prod:
                max_prod = new_prod
                print("Found new max \\ diagonal", max_prod)
            
print("Final answer is", max_prod)
```
