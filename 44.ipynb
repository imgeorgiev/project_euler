Pentagonal numbers are generated by the formula, $P_n=\dfrac{n(3n−1)}{2}$. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k − P_j|$ is minimised; what is the value of D?

```python
# generate a bunch of pentagonal numbers
pentagonal = []
for n in range(1,10000):
    pentagonal.append(n*(3*n-1)/2)
```

```python
from math import *

def is_pentagonal(P):
    if (1 + sqrt(1+4*3*2*P))/6 % 1 == 0:
        return True
    else:
        return False
        
is_pentagonal(92)
```

```python
hits = []
for i in range(len(pentagonal)):
    for j in range(len(pentagonal)):
        if is_pentagonal(pentagonal[i] + pentagonal[j]):
            if pentagonal[j] > pentagonal[i]:
                if is_pentagonal(pentagonal[j] - pentagonal[i]):
                    hits.append((i,j))
            else:
                if is_pentagonal(pentagonal[i] - pentagonal[j]):
                    hits.append((i,j))
```

```python
# now find the smallest difference
target = 1000000000000000
for i,j in hits:
    if abs(pentagonal[j] - pentagonal[i]) < target:
        print(i,j)
        target = abs(pentagonal[j] - pentagonal[i])
target
```
